# Notebook for crawling OpenAlex institutions

```python
import os, time, requests, re, json, random, csv
```

```python
OPENALEX_WORK_API = "https://api.openalex.org/institutions"

def get_openalex_paper_info(instname):
    query_url = f'{OPENALEX_WORK_API}?search="{instname}"&select=id,display_name,country_code,type,homepage_url'
#     print(query_url)
    response = requests.get(query_url)

    if response.status_code == 200:
        data = response.json()
        match_count = data["meta"]["count"]
        matched = None
        if match_count == 0:
            return None
        if match_count > 1:
            for m in data["results"]:
                if m["display_name"] == instname:
                    matched = m
                    break
        else:
            matched = data["results"][0]
        return matched
    else:
        return None
```

```python
DIR_RAW_DATA = os.path.join("../../data/scores")

instName = set()

def readScores():
    try:
        for cfile in os.listdir(DIR_RAW_DATA):
            cflist = json.load(open(os.path.join(DIR_RAW_DATA, cfile), "r"))
            for k, v in cflist.items():
                instName.add(k)
    except Exception as e:
        print(e)
```

```python
readScores()
print(len(instName))
# print(instName)
```

```python
type_set = set()
m_count = 0
with open('inst_set.csv', 'a', newline='') as csvfile:
    spamwriter = csv.writer(csvfile)
    for i, inst in enumerate(list(instName)):
        if inst == "":
            continue
        m = get_openalex_paper_info(inst)
        if m == None:
            continue
        m_count += 1
        print("{}/{} {} {}".format(m_count, i, inst, m["id"]))
        type_set.add(m["type"])
        spamwriter.writerow([m["id"], m["display_name"], m["type"], m["country_code"], m["homepage_url"]])
```
